    # Simple Add-Delete Morphology Implementation

    # Add-Delete Rules (suffix based)
    rules = [
        ("ing", ""),   # playing → play
        ("ed", ""),    # played → play
        ("s", ""),     # plays → play
        ("ies", "y"),  # studies → study
        ("ness", ""),  # happiness → happy
        ("ment", ""),  # development → develop
    ]

    def apply_rules(word):
        for delete, add in rules:
            if word.endswith(delete):
                base = word[:-len(delete)] + add
                return {
                    "original": word,
                    "delete": delete,
                    "add": add,
                    "root": base
                }
        return {
            "original": word,
            "delete": "-",
            "add": "-",
            "root": word
        }

    # Sample words
    words = ["playing", "played", "plays", "studies", "happiness", "development", "cat"]

    # Apply morphology
    results = [apply_rules(word) for word in words]

    # Print Add-Delete Table
    print("{:<12} {:<10} {:<10} {:<12}".format("Word", "Delete", "Add", "Root"))
    print("-"*45)

    for r in results:
        print("{:<12} {:<10} {:<10} {:<12}".format(
            r["original"], r["delete"], r["add"], r["root"]
        ))
